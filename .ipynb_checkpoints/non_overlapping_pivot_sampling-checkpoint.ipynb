```python
# now running many experiments and doing analyses on them all

from neuron import h
from neuron.units import mV, ms
h.load_file('stdrun.hoc')

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

import numpy as np
import pandas as pd
import random
import math
import pickle
```

```python
def gen_histories(sim_length = 1000000, stim_interval = 10, num_histories = 1000):

    # Set up biophysical model
    axon = h.Section(name='axon')
    axon.insert(h.hh)

    # add a synapse
    syn = h.ExpSyn(axon(0))
    syn.tau = 1 * ms
    syn.e = 0 * mV
    syn_current = h.Vector().record(syn._ref_i)

    # add a stimulus
    stim = h.NetStim()
    stim.number = 9999999
    stim.interval = stim_interval * ms
    stim.noise = True
    stim.start = 0 * ms

    stim_times = h.Vector()

    # connect stimulus to synapse
    nc = h.NetCon(stim, syn)
    nc.delay = 0 * ms
    nc.weight[0] = 0.2
    nc.record(stim_times)

    # setup recording
    _t = h.Vector().record(h._ref_t)
    _v = h.Vector().record(axon(0.5)._ref_v)
    _m = h.Vector().record(axon(0.5).hh._ref_m)
    _n = h.Vector().record(axon(0.5).hh._ref_n)
    _h = h.Vector().record(axon(0.5).hh._ref_h)
    spike_times = h.Vector()
    nc_self = h.NetCon(axon(0.5)._ref_v, None, sec=axon)
    nc_self.record(spike_times)

    # run simulation
    h.finitialize(-65 * mV)
    h.continuerun(sim_length * ms)
    
    histories = np.array((_v, _m, _h, _n))[:, random.sample(range(0, len(_v)), num_histories)]

    return histories, np.array((_t, _v, _m, _h, _n, syn_current)), list(spike_times), list(stim_times)
```

```python
# set up preceding stim times for event horizon experiment (based on output spike)

def center_preceding_stims(spike, stims): 
    spike = (math.floor((spike - stims[-1]) * 40))/40
    stims = [(math.floor((stim - stims[-1]) * 40))/40 for stim in stims]
    
    return spike, stims

def view_preceding_stimuli(centered_stim_set, centered_spike, centered_spike_sets):
    # view the preceding stimuli
    plt.figure(figsize = (15, 3))
    plt.vlines(centered_stim_set, 0, 1, color = 'black')
    plt.vlines(centered_spike, 0, 1, color = 'red')
    plt.vlines(centered_spike_sets, 0, 1, color = 'limegreen')
    
    plt.xlabel('time (ms)')
    plt.yticks([])
    plt.show()
    
    return
```

```python
def gen_non_overlapping_preceding_stim_sets(stim_times, spike_times, num_preceding_stimuli):
    
    # get all possible sampled spikes
    possible_sampled_spikes={
        i:[] for i, spike in enumerate(spike_times) if spike > stim_times[num_preceding_stimuli-1]
    }
    
    # get the preceding stims for each possible sampled spike
    for spike in possible_sampled_spikes:
        preceding_stims = [stim for stim in stim_times if stim < spike_times[spike]]
        preceding_stims = preceding_stims[-num_preceding_stimuli:]
        possible_sampled_spikes[spike] = preceding_stims
    
    used_spike_inds = [] # hold the sampled spikes indices that do not overlap
    used_windows = [] # hold the used windows (time spans of sampled spike and stim sets, to troubleshoot) 
    used_stim_inds = [] # hold the stimuli used for sampled spikes
    
    centered_spikes = []
    centered_stim_sets = []
    centered_spike_sets = []
    
    while len(possible_sampled_spikes) > 0:
        current_spike = random.choice(list(possible_sampled_spikes.keys()))
        
        if not bool(set(possible_sampled_spikes[current_spike]) & set(used_stim_inds)):
            # if this sampled spike's stimuli are already used, then toss it
            # otherwise we keep it
            used_stim_inds += possible_sampled_spikes[current_spike]
            used_spike_inds.append(current_spike)
            
            # output this set's window to check if it worked correctly
            window_start = possible_sampled_spikes[current_spike][0]
            window_end = spike_times[current_spike]
            used_windows.append([window_start, window_end])
            
            # center spikes and thier preceding stims sets around the pivot
            centered_spike, centered_stim_set = center_preceding_stims(spike_times[current_spike], possible_sampled_spikes[current_spike])
            centered_spikes.append(centered_spike)
            centered_stim_sets.append(centered_stim_set)
            
            # gather spikes withing this window and center them as well
            centered_spike_set = [spike for spike in spike_times if spike > window_start and spike < window_end]
            centered_spike_set = [(math.floor((spike - possible_sampled_spikes[current_spike][-1]) * 40))/40 for spike in centered_spike_set]
            centered_spike_sets.append(centered_spike_set)
        
        del possible_sampled_spikes[current_spike]
        
    #print('fraction of sampled spikes: ', len(used_spike_inds), '/', len(spike_times), ' - ',len(used_spike_inds)/len(spike_times))
    
    return used_spike_inds, used_windows, centered_spikes, centered_stim_sets, centered_spike_sets

used_spikes, used_windows, centered_spikes, centered_stim_sets, centered_spike_sets = gen_non_overlapping_preceding_stim_sets(
    stim_times, spike_times_initial, 20)
```

```python
# see how this number increases with multiple simulation time and num_preceding_stims
num_used_spikes = []
num_spikes_total = []
used_spike_ratio = []
num_preceding_stims = []

for n in range(10,32,2):
    used_spikes, used_windows, centered_spikes, centered_stim_sets, centered_spike_sets = gen_non_overlapping_preceding_stim_sets(
        stim_times, spike_times_initial, n)
    num_used_spikes.append(len(used_spikes))
    num_spikes_total.append(len(spike_times_initial))
    used_spike_ratio.append(len(used_spikes)/len(spike_times_initial))
    num_preceding_stims.append(n)
    
plt.plot(num_preceding_stims, used_spike_ratio)
plt.xlabel('number of preceding stims')
plt.ylabel('used spike ratio')
plt.show()
```
